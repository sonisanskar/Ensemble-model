```python
algo=['Linear Regression',
'Ridge Regression',
'Lasso',
'SVR (Linear)',
'SVR (rbf)',
'Decision Tree',
'Random Forest',
'Elastic Net',
'Gradient Boost',
]
```

```python
l=['base11.sav','base22.sav','base33.sav','base44.sav','base5.sav','base66.sav','base77.sav','base88.sav','base99.sav']
#these are all the base models with their names corresponding to algos respectively
```

```python
def make_bases(df):
    X=df.drop(['Variables','PX_HIGH'],axis=1)
    y=df['PX_HIGH']
    start={'PX_HIGH':y}
    pred=pd.DataFrame(start)
    mae=[]
    mse=[]
    rmse=[]
    score=[]
    metric=pd.DataFrame(index=algo)
    row=0
    for i in l:
        
        loaded_model = pickle.load(open(i, 'rb'))     #loading the model one by one which are named  according to the elements in list l
        result = loaded_model.score(X, y)
        
        prediction=loaded_model.predict(X)
        print(result,i,r2_score(y,prediction))
        pred[i]=prediction
        print('MAE:', metrics.mean_absolute_error(y, prediction))
        print('MSE:', metrics.mean_squared_error(y, prediction))
        print('RMSE:', np.sqrt(metrics.mean_squared_error(y, prediction)))
        mae.append(metrics.mean_absolute_error(y, prediction))
        mse.append(metrics.mean_squared_error(y, prediction))
        rmse.append(np.sqrt(metrics.mean_squared_error(y, prediction)))
        score.append(r2_score(y,prediction))
        
    data={'models':algo,"RMSE":rmse,"R_2SCORE":score,"MAE":mae,'MSE':mse,}
    metric = pd.DataFrame(data)#create a dataframe including the metrics from all base models
    
        
        
  
         
        
    return pred,metric
        
        
    
    
```

```python
datas,metric=make_bases(df)
```

```python
def final_model(datas):
    datas=datas.rename(columns={'base1.sav':'model_1','base22.sav':'model_2','base33.sav':'model_3','base44.sav':'model_4','base5.sav':'model_5','base66.sav':'model_6','base77.sav':'model_7','base88.sav':'model_8','base99.sav':'model_9'})
    X=datas.drop(['model_9','PX_HIGH','model_5','model_6','model_7'],axis=1)
    y=datas['PX_HIGH']
    loaded_model = pickle.load(open('top44.sav', 'rb'))
    result = loaded_model.score(X, y)    
    prediction=loaded_model.predict(X)
    mae=[]
    mse=[]
    rmse=[]
    score=[]
    metric=pd.DataFrame(index=l1)
    print(result,r2_score(y,prediction))
    
    print('MAE:', metrics.mean_absolute_error(y, prediction))
    print('MSE:', metrics.mean_squared_error(y, prediction))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y, prediction)))
    mae.append(metrics.mean_absolute_error(y, prediction))
    mse.append(metrics.mean_squared_error(y, prediction))
    rmse.append(np.sqrt(metrics.mean_squared_error(y, prediction)))
    score.append(r2_score(y,prediction))
    
    data={"models":l1,"RMSE":rmse,"R_2SCORE":score,"MAE":mae,'MSE':mse}
    metric = pd.DataFrame(data) 
    return metric
        
        
        
    
    
```

```python
pro_metric=final_model(datas) #returning the datafrme  containing metrics of the proposed model
```

```python
#appending both the dataframe (containing accuracies of base and proposed model)
last=metric.append(pro_metric)
```
